```python
# Finding Length of stay of patients
```

```python
df['Date of Admission'] = pd.to_datetime(df['Date of Admission'], format='%Y-%m-%d')
df['Discharge Date'] = pd.to_datetime(df['Discharge Date'], format='%Y-%m-%d')
df['Length of Stay'] = (df['Discharge Date'] - df['Date of Admission']).dt.days
```

```python
print(df)
```

```python
# Avg Length of stay:
```

```python
average_length_of_stay = df['Length of Stay'].mean()
```

```python
plt.hist(df['Length of Stay'], bins=10, color='green', edgecolor='black', alpha=0.7)
plt.axvline(average_length_of_stay, color='red', linestyle='dashed', linewidth=2, label='Avg Length of Stay')
plt.xlabel('Length of Stay (days)')
plt.ylabel('Frequency')
plt.title('Distribution of Length of Stay')
plt.legend()
plt.show()
```
